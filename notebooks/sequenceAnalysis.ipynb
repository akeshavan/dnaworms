```python
import numpy as np
import re
```

```python
dna = ["A","G","C","T"]
```

```python
# define total length of scaffold
L = 700

# define scaffold (will be an input later)
scaf = open("m13.txt").read()#''.join([dna[x] for x in np.random.random_integers(0,3,L)])
```

```python
def get_motif(foo,L,start):
    if L+start <= len(foo):
        return foo[start:L+start]
    return ''
```

```python
def count_motif(foo,motif):
    idx = [m.start() for m in re.finditer('(?=%s)'%motif, foo)]
    return idx
```

```python
def plot_bar(vals,**kwargs):
    names = [v[0] for v in vals]
    numbers = [v[1] for v in vals]
    N = xrange(len(numbers))
    width = 0.35 
    fig, ax = plt.subplots(**kwargs)
    ax.bar(N,numbers,width=width)
    ax.set_xticks([n+width/2 for n in N])
    ax.set_xticklabels(names,rotation=90);
    return ax
```

```python
counts = []
for x in range(20,3,-1):
    motif = np.unique(np.array([get_motif(scaf,x,p) for p in xrange(L)]))
    count = [[m]+count_motif(scaf,m) for m in motif if m]
    counts.append(count)
```
