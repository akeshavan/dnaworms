# Noodle Generator

* We want to automatically generate noodle designs with variable flexibilities. 
* This means we need to try different crossover/kink motifs of the scaffold and staple strands.
* The first step is to define locations where crossovers can happen

## Pre-Crossovers

Here is our cross section:

```python
#Image("cadnano2001.png")
```

```python
import networkx as nx
G=nx.Graph()
G.add_nodes_from(range(6))
G.add_edge(0,1,label="0 + 21n",start_idx=0)
G.add_edge(1,2,label="14 + 21n",start_idx=14)
G.add_edge(2,3,label="7 + 21n",start_idx=7)
G.add_edge(3,4,label="0 + 21n",start_idx=0)
G.add_edge(4,5,label="14 +21n",start_idx=14)
G.add_edge(0,5,label="7 + 21n",start_idx=7)
```

```python
nx.draw_circular(G,node_color="lightyellow",node_size=1000)
pos=nx.circular_layout(G)
edge_labels=dict([((u,v,),d['label'])
             for u,v,d in G.edges(data=True)])
nx.draw_networkx_edge_labels(G,pos,edge_labels=edge_labels);
title("Crossover Locations for Staples");
```

#### Initialize our helices:

```python
length = 84
length +=1
helices = {}
for i in range(6):
    helices[i]= -np.ones(length)
```

```python
def plot_helices(helices):
    figure()
    hold(True)
    for i,data in helices.iteritems():
        plot([0,length],[i,i])
    
    axis([-2,length,-1,6])
    yticks(range(6))
    xticks(range(0,length,7))
plot_helices(helices)
```

#### Define crossover locations for each Edge on helices

```python
for u,v,d in G.edges(data=True):
    N = (length-d["start_idx"])/21
    idx = [d["start_idx"]+21*n for n in range(N+1)]
    helices[u][idx] = v 
    helices[v][idx] = u 
```

```python
def plot_helices(helices):
    figure()
    hold(True)
    for i,data in helices.iteritems():
        plot([0,length],[i,i])
        idx = np.nonzero(data !=-1)[0]
        y = data[idx]
        plot(idx,y,'o',color="grey",alpha=0.5)
        
        for ii,d in enumerate(y):
            annotate('%d'%d,[idx[ii],i])
        
        
    axis([-2,length,-1,6])
    yticks(range(6))
    xticks(range(0,length,7))
    
plot_helices(helices)
```

## Crossovers

* We need to define a new scheme for when crossovers actually happen!
* Lets make it similar to the helices structure:

```python
Xs = {}
for i in range(6):
    Xs[i]= zeros(length).astype(bool)
```

* In this case, Xs is a boolean, and the index of a True matches to the index of helices. 
* Now we can easily validate if a structure is valid if there is not a -1 in helices[Xs]

```python
def validateX(helices,Xs):
    for helix, data in helices.iteritems():
        if -1 in data[Xs[helix]]:
            raise Exception("Invalid!")
        #TODO: add in part for reciprocity
    return True
```

```python
def plot_XKhelices(helices,Xs,Ks):
    figure(figsize=(12,4))
    hold(True)
    for i,data in helices.iteritems():
        plot([0,length],[i,i],color="b")
        idx = np.nonzero(data !=-1)[0]
        y = data[idx]
        plot(idx,y,'o',color="grey",alpha=0.5)
        
        for ii,d in enumerate(y):
            annotate('%d'%d,[idx[ii],i])
        
        Xdata = Xs[i]
        Xidx = np.nonzero(Xdata)[0] # <-- list of indices where crossovers happen!
        for Xi in Xidx:
            plot([Xi,Xi],[i,data[Xi]])
        
        Kdata = Ks[i]
        Kidx = np.nonzero(Kdata)[0]
        Ky = ones(len(Kidx))*i
        plot(Kidx,Ky,'x',color="black")
        
    axis([-2,length,-1,6])
    yticks(range(6))
    xticks(range(0,length,7))
    
plot_XKhelices(helices,Xs,Ks)
```

We have way too many kinks! The staples are too short. We need to validate kinks better

```python
def validateK(Xs,Ks):
    for helix, data in Xs.iteritems():
        if -1 in data[Ks[helix]]:
            raise Exception("Invalid!")
        kidx = np.nonzero(Ks[helix])[0]
        kdist = np.diff(kidx)
        if (kdist<6).any():
            raise(Exception("Kinks are too close together!"))
    return True
validateK(Xs,Ks)
```

Format
-------

Now we need to convert our structures to valid CadNano formatting with arrays of 4 items:
[Helix before, position before, Helix after, position after]

* We first need to define the directions (5' --> 3', 3' -->5') of each helix
* Then we generate the 4some in the correct order, stopping at crossovers or kinks
* Then we'll push it to a cadnano json and see if it works.


```python
G.node[0]["direction"]=3
G.node[2]["direction"]=3
G.node[4]["direction"]=3
G.node[1]["direction"]=5
G.node[3]["direction"]=5
G.node[5]["direction"]=5
```

```python
def init_cadnano(helix, direction, length):
    foo = []
    for i in range(length):
        if direction == 5:
            foo.append([helix,i-1,helix,i+1])
        elif direction == 3:
            foo.append([helix,i+1,helix,i-1])
        if i==0:
            if direction==5: foo[i] = [-1,-1,helix,i+1]
            elif direction==3: foo[i] = [helix,i+1,-1,-1]
        elif i==length-2:
            if direction==5: foo[i] = [helix,i-1,-1,-1]
            elif direction==3: foo[i] = [-1,-1,helix,i-1]
    return foo
```

```python
class Noodle(object):
    def __init__(self,length=84):
        self._length = length+1
        self._init_helices()
        self._init_crossovers()
        self._init_kinks()
        
    def _init_helices(self):    
        self.helices = {}
        for i in range(6):
            self.helices[i]= -np.ones(self._length)
        self._helix_graph()
        for u,v,d in G.edges(data=True):
            N = (self._length-d["start_idx"])/21
            idx = np.asarray([d["start_idx"]+21*n for n in range(N+1)])
            self.helices[u][idx] = v 
            self.helices[v][idx] = u           
            self.helices[u][idx[np.nonzero(idx)]-1] = v 
            self.helices[v][idx[np.nonzero(idx)]-1] = u 

    def _helix_graph(self):
        G=nx.Graph()
        G.add_nodes_from(range(6))
        G.node[0]["direction"]=3
        G.node[2]["direction"]=3
        G.node[4]["direction"]=3
        G.node[1]["direction"]=5
        G.node[3]["direction"]=5
        G.node[5]["direction"]=5
        G.add_edge(0,1,label="0 + 21n",start_idx=0)
        G.add_edge(1,2,label="14 + 21n",start_idx=14)
        G.add_edge(2,3,label="7 + 21n",start_idx=7)
        G.add_edge(3,4,label="0 + 21n",start_idx=0)
        G.add_edge(4,5,label="14 +21n",start_idx=14)
        G.add_edge(0,5,label="7 + 21n",start_idx=7)
        self.G = G
        
    def draw_helix_graph(self):
        nx.draw_circular(self.G,node_color="lightyellow",node_size=1000)
        pos=nx.circular_layout(self.G)
        edge_labels=dict([((u,v,),d['label'])
                     for u,v,d in self.G.edges(data=True)])
        nx.draw_networkx_edge_labels(self.G,pos,edge_labels=edge_labels);
        title("Crossover Locations for Staples");
        
    def _init_crossovers(self):
        self.Xs = {}
        for i in range(6):
            self.Xs[i]= zeros(length).astype(bool)
    
    def validate_crossovers(self):
        for helix, data in self.helices.iteritems():
            if -1 in data[self.Xs[helix]]:
                raise Exception("Invalid!")
            #TODO: add in part for reciprocity
        return True
    
    def _init_kinks(self):
        self.Ks = {}
        for i in range(6):
            self.Ks[i]= zeros(length).astype(bool)
            
    def validate_kinks(self):
        for helix, data in self.Xs.iteritems():
            if -1 in data[self.Ks[helix]]:
                raise Exception("Invalid!")
            kidx = np.nonzero(self.Ks[helix])[0]
            kdist = np.diff(kidx)
            if (kdist<6).any():
                raise(Exception("Kinks are too close together!"))
        return True
            
    def _init_cadnano(self,helix, direction, length):
        foo = []
        for i in range(self._length):
            if direction == 5:
                foo.append([helix,i-1,helix,i+1])
            elif direction == 3:
                foo.append([helix,i+1,helix,i-1])
            if i==0:
                if direction==5: foo[i] = [-1,-1,helix,i+1]
                elif direction==3: foo[i] = [helix,i+1,-1,-1]
            elif i==length-2:
                if direction==5: foo[i] = [helix,i-1,-1,-1]
                elif direction==3: foo[i] = [-1,-1,helix,i-1]
        return foo
        
    def _draw_Xovers_and_kinks(self,cad,direction,X,K,data,helix):
        Xidx = np.nonzero(X)[0]
        Kidx = np.nonzero(K)[0]
        
        for idx in Xidx:
            if direction==5 and not idx%7: cad[idx] = [int(data[idx]),idx,helix,idx+1];
            elif direction==5 and idx%7: cad[idx] = [helix,idx-1,int(data[idx]),idx]; 
            
            elif direction==3 and idx%7: cad[idx] = [int(data[idx]),idx,helix,idx-1]; 
            elif direction==3 and not idx%7: cad[idx] = [helix,idx+1,int(data[idx]),idx]; 
        
        for idx in Kidx:
            if direction==5: cad[idx] = [helix,idx-1,-1,-1]
            elif direction==3: cad[idx] = [-1,-1,helix,idx-1]
        
        return cad
        
    def to_cadnano(self,outfile="noodle.json"):
        self.stap = {}
        for helix, d in self.G.nodes(data=True):
            cad = self._init_cadnano(helix,d['direction'],self._length)
            X = self.Xs[helix][:-1]
            K = self.Ks[helix][:-1]
            data = self.helices[helix][:-1]
            stap[helix] = self._draw_Xovers_and_kinks(cad,d['direction'],X,K,data,helix)
        json = load_json('template.json')
        for i,Helix in enumerate(json['vstrands']):
            for key,item in Helix.iteritems():
                # Initialize everyting to the right length: skip, scaf, loop
                Helix['skip'] = [[0]]*(length-1) 
                Helix['loop'] = [[0]]*(length-1) 
                Helix['scaf'] = [[-1,-1,-1,-1]]*(length-1) 
                Helix['stap'] = stap[i][:-1]
        save_json(outfile,json)
        
    def plot(self):
        figure(figsize=(12,4))
        hold(True)
        for i,data in self.helices.iteritems():
            plot([0,self._length],[i,i],color="b")
            idx = np.nonzero(data !=-1)[0]
            y = data[idx]
            plot(idx,y,'o',color="grey",alpha=0.5)
            
            for ii,d in enumerate(y):
                annotate('%d'%d,[idx[ii],i])
            
            Xdata = self.Xs[i]
            Xidx = np.nonzero(Xdata)[0] # <-- list of indices where crossovers happen!
            for Xi in Xidx:
                plot([Xi,Xi],[i,data[Xi]])
            
            Kdata = self.Ks[i]
            Kidx = np.nonzero(Kdata)[0]
            Ky = ones(len(Kidx))*i
            plot(Kidx,Ky,'x',color="black")
            
        axis([-2,self._length,-1,6])
        yticks(range(6))
        xticks(range(0,self._length,7))
        
    def randomize(self):
        for helix, data in self.helices.iteritems():
            idx = np.nonzero(data !=-1)[0]
            self.Xs[helix][idx] = np.random.random_integers(0,1,len(idx)).astype(bool) 

        for helix,data in self.helices.iteritems():
            neighbors = np.unique(data)[1:].astype(int)
            for n in neighbors:
                idx = np.nonzero(data.astype(int)==n)[0]
                self.Xs[n][idx] += self.Xs[helix][idx]
 
        for helix, data in helices.iteritems():
            Xdata = self.Xs[helix]
            self.Ks[helix][np.multiply(Xdata ==False,data!=-1)] = True
```
