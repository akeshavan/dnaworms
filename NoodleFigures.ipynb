```python
#Made Up Results
CirculationTime = np.random.rand(6) #circulation times
vectors = np.random.rand(6,4)
row_sums = vectors.sum(axis=1)
Retention = vectors / row_sums[:, numpy.newaxis] #% retentions
Rnames = ['Tumor','Liver','Spleen','Other']
```

```python
def plot_bar(names,vals,ax,labels=False):
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    N = len(vals)
    width = 0.35
    ax.barh(range(1,N+1),vals,color=colors)
    ax.set_xticks([0,0.5,1])
    ax.set_xticklabels([0,0.5,1])
    if not labels:
        ax.set_yticklabels(names)
        ax.set_yticks([width + i for i in range(1,N+1)])
    else:
        ax.set_yticks([])
```

```python
def plot_noodle(ax,idx,cd=0):
    ax.plot(linspace(0,2*pi),sin(linspace(0,2*pi)*(idx+1))*(idx+1)/3,'limegreen',linewidth=5)
    if cd:
        ax.plot(linspace(0,2*pi,5*cd),sin(linspace(0,2*pi,5*cd)*(idx+1))*(idx+1)/3,'ro')
    ax.set_axes([0,2*pi,0,1])
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
    ax.axis('off')
```

```python
def plot_pie(names,sizes,ax):
    colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
    explode = (0, 0, 0, 0.1) # only "explode" the 2nd slice (i.e. 'Hogs')
    
    ax.pie(sizes, explode=explode, labels=names, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=90)
    # Set aspect ratio to be equal so that pie is drawn as a circle.
    #axis('equal')
```

```python
fig = plt.figure(figsize=(18,6))
gs = GridSpec(2,6, height_ratios=[1, 3]) 
for i in range(6):
    ax = plt.subplot(gs[1,i])
    plot_bar(Rnames[::-1],Retention[i,:],ax,i)
    cdnum = i%3
    ax = plt.subplot(gs[0,i])
    plot_noodle(ax,[0,0,0,1,1,1][i],cdnum)
    ax.set_title(Noodles[i])
#plot_noodle(ax,i,cd=0)
fig.text(0.5, 0.04, 'Fraction of Nanoparticles in Region', ha='center', va='center')    
fig.text(0.5, 1.04, 'Biodistribution of Nanostructures', ha='center', va='center',fontsize=18);    
savefig("img/Bar_Biodistribution.png")
```
